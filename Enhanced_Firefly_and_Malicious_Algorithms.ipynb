**FIREFLY ALGORITH COMPARISON WITH DL, SEIR-F, AND EFF-FAODV**

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("/content/Firefly_dataset.csv")

df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df["Nodes"], df["DL_Time(ms)"], label="DL", marker='o')
plt.plot(df["Nodes"], df["SEIR-F_Time(ms)"], label="SEIR-F", marker='s')
plt.plot(df["Nodes"], df["EFF-FAODV_Time(ms)"], label="EFF-FAODV", marker='^')

plt.xlabel("Number of Nodes")
plt.ylabel("Time Taken (milliseconds)")
plt.title("Performance Comparison: DL vs. SEIR-F vs. EFF-FAODV")
plt.legend()
plt.grid(True)
plt.show()
```

**% OF THROUGHPUT OF FIREFLY ALGORITHM GRAPH**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = pd.read_csv("/content/Throughput_firefly_Sample.csv")

df = pd.DataFrame(data)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(df["Nodes"], df["DL_Throughput(%)"], label="DL", marker='o', linestyle='--')
plt.plot(df["Nodes"], df["SEIR-F_Throughput(%)"], label="SEIR-F", marker='s', linestyle='-.')
plt.plot(df["Nodes"], df["EFF-FAODV_Throughput(%)"], label="EFF-FAODV", marker='^', linestyle='-')

plt.xlabel("Number of Nodes")
plt.ylabel("Throughput (%)")
plt.title("Throughput Comparison: DL vs. SEIR-F vs. EFF-FAODV")
plt.legend()
plt.grid(True)
plt.ylim(0, 100)  # Throughput ranges from 0% to 100%
plt.show()
```

**COMPARISONS BASED ON ENERGY CONSUMPTION using bar chart**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Data
data = pd.read_csv("/content/Energy_consumption_EFF.csv")

df = pd.DataFrame(data)

# Plot settings
x = np.arange(len(df["Nodes"]))  # Node counts as x-axis
width = 0.25  # Bar width

fig, ax = plt.subplots(figsize=(12, 6))
rects1 = ax.bar(x - width, df["DL_Energy(ms)"], width, label="DL", color="red", edgecolor="black")
rects2 = ax.bar(x, df["SEIR-F_Energy(ms)"], width, label="SEIR-F", color="orange", edgecolor="black")
rects3 = ax.bar(x + width, df["EFF-FAODV_Energy(ms)"], width, label="EFF-FAODV", color="green", edgecolor="black")

# Labels and titles
ax.set_xlabel("Number of Nodes")
ax.set_ylabel("Energy Consumption (% of time in ms)")
ax.set_title("Energy Consumption Comparison: DL vs. SEIR-F vs. EFF-FAODV")
ax.set_xticks(x)
ax.set_xticklabels(df["Nodes"])
ax.legend()
ax.grid(axis="y", linestyle="--")

plt.ylim(0, 140)  # Adjust y-axis range
plt.show()
```

**MALICIOUS NODE DETECTION **

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = pd.read_csv("/content/Malicious_node_detection.csv")

df = pd.DataFrame(data)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(df["Nodes"], df["DL_Detected"], label="DL", marker="o", linestyle="--", color="red")
plt.plot(df["Nodes"], df["SEIR-F_Detected"], label="SEIR-F", marker="s", linestyle="-.", color="blue")
plt.plot(df["Nodes"], df["EFF-FAODV_Detected"], label="EFF-FAODV", marker="^", linestyle="-", color="green")
plt.plot(df["Nodes"], df["Malicious_Nodes"], label="Actual Malicious Nodes", linestyle=":", color="black")

# Formatting
plt.xlabel("Number of Nodes")
plt.ylabel("Malicious Nodes Detected")
plt.title("Malicious Node Detection: DL vs. SEIR-F vs. EFF-FAODV")
plt.legend()
plt.grid(True)
plt.xticks(df["Nodes"])
plt.show()
```

**PACKET DELIVERY RATION BASED ON MALICIOUS DATA PACKET RATIO**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = pd.read_csv("/content/Packet_delivery_Malicious.csv")

df = pd.DataFrame(data)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(df["Nodes"], df["DL_PDR(%)"], label="DL", marker="o", linestyle="--", color="red")
plt.plot(df["Nodes"], df["SEIR-F_PDR(%)"], label="SEIR-F", marker="s", linestyle="-.", color="blue")
plt.plot(df["Nodes"], df["EFF-FAODV_PDR(%)"], label="EFF-FAODV", marker="^", linestyle="-", color="green")

# Formatting
plt.xlabel("Number of Nodes")
plt.ylabel("Packet Delivery Ratio (%)")
plt.title("Packet Delivery Ratio Comparison: DL vs. SEIR-F vs. EFF-FAODV")
plt.legend()
plt.grid(True)
plt.xticks(df["Nodes"])
plt.ylim(40, 100)  # PDR ranges from 0% to 100%
plt.show()
```
